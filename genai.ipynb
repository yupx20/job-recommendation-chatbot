```python
from google import genai
from google.genai import types
from IPython.display import Markdown

client = genai.Client(api_key="Enter Google Gemini API key here")

MODEL_ID= 'gemini-2.5-flash'
response= client.models.generate_content(
  model=MODEL_ID,
  contents=[
    'Explain how AI works.',
  ]
)

# print(response.text)
Markdown(response.text)
```

```python
response = client.models.generate_content_stream(
    model="gemini-2.5-flash", contents="Write a story about king Arthur."
)

for chunk in response: # Untuk menampilkan hasil generate secara parsial
    print(chunk.text)
    print("=" * 100)
```

```python
from google.genai import types

doc = "Text_Analysis_Based_on_Natural_Language_Processing_NLP.pdf"

with open(doc, 'rb') as f:
  doc_data = f.read()

prompt = "Summarize this document"
response = client.models.generate_content(
  model="gemini-2.5-flash",
  contents=[
      types.Part.from_bytes(
        data=doc_data,
        mime_type='application/pdf',
      ),
      prompt]
)

print(response.text)
```
